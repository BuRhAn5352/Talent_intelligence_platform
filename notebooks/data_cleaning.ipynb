```python
import re

def ultimate_clean(text):
    if not isinstance(text, str): return ""
    
    # PROTECT TECH TERMS (Prevents regex from killing symbols in C++, C#, .NET)
    text = text.replace('C++', 'CPP_TERM').replace('C#', 'CSHARP_TERM').replace('.NET', 'DOTNET_TERM')

    # REMOVE EMAILS, LINKS, AND PHONE NUMBERS
    text = re.sub(r'\S+@\S+', ' ', text)  # Emails
    text = re.sub(r'http\S+|www\S+', ' ', text)  # URLs
    text = re.sub(r'\d{10,}', ' ', text)  # Phone numbers (10+ digits)

    # 3. REMOVE NAUKRI-SPECIFIC NOISE (Boilerplate phrases)
    noise_patterns = [
        r'Job Description', r'Send me Jobs like this', r'Not Disclosed by Recruiter',
        r'Greetings from', r'Please share your Resume on', r'Note :', r'Pay scale is open',
        r'Download', r'PPT', r'Photo', r'Desired Candidate Profile', r'Company Profile'
    ]
    for pattern in noise_patterns:
        text = re.sub(pattern, ' ', text, flags=re.IGNORECASE)

    # 4. REMOVE SPECIAL CHARS (But keep spaces so words don't merge)
    # This turns "PHP/MySQL" into "PHP MySQL"
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # 5. RESTORE TECH TERMS
    text = text.replace('CPP_TERM', 'C++').replace('CSHARP_TERM', 'C#').replace('DOTNET_TERM', '.NET')

    # 6. STRIP EXTRA WHITESPACE
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

#  STEPS 

# A. Apply the main cleaning function
df['cleaned_description'] = df['jobdescription'].apply(ultimate_clean)

# Remove "View Contact Details" junk and Leading numbers
# These often survive the function because they are mixed with numbers
df['cleaned_description'] = df['cleaned_description'].str.replace(r'\d+\s*view contact details.*', '', flags=re.IGNORECASE, regex=True)
df['cleaned_description'] = df['cleaned_description'].str.replace(r'^\d+[\.\s]*', '', regex=True)

#  Remove "INR" and Salary blocks
df['cleaned_description'] = df['cleaned_description'].str.replace(r'INR\s*[\d,]*\s*[\d,]*\s*P\.A', '', flags=re.IGNORECASE, regex=True)

print("Cleaning Finished. Sample of Row 0:")
print(df['cleaned_description'].iloc[0][:500])
```

```python
def clean_skills_category(text):
    if not isinstance(text, str): return "general"
    # 1. Remove "IT Software - " prefix (it's redundant in 80% of rows)
    text = text.replace('IT Software - ', '')
    # 2. Replace special chars with spaces
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    # 3. Clean whitespace
    return re.sub(r'\s+', ' ', text).strip().lower()

df['skills_cleaned'] = df['skills'].apply(clean_skills_category)
```
